```python
first_neighbor = pd.read_csv('query-result_1_n_50000.csv')
first_neighbor
```

```python
print((first_neighbor.iloc[0,0]).split("/")[-1])
```

```python
for index in first_neighbor.index:
    first_neighbor.loc[index, 'x'] = (first_neighbor.iloc[index,0]).split("/")[-1]
```

```python
first_neighbor
```

```python
for index in first_neighbor.index:
    first_neighbor.loc[index, 'y'] = (first_neighbor.iloc[index,1]).split("/")[-1]
```

```python
first_neighbor
```

```python
drug_interactions = first_neighbor[['x', 'y']]
drug_interactions
```

```python
drug_interactions.columns = ['drug', 'target']
drug_interactions
```

```python
drug_interactions = drug_interactions.drop_duplicates()
drug_interactions
```

```python
rel_interaction = []
for i in range(len(drug_interactions)):
    rel_interaction.append('interactsWith')
print(len(rel_interaction))
```

```python
drug_interactions.insert(1, 'relation', rel_interaction)
```

```python
drug_interactions
```

```python
ff_data = pd.concat([yamanishi, drug_interactions])
ff_data
```

```python
ff_data.to_csv('ff_data.txt', header=False, index=False)
```

```python
ff_data = ff_data.drop(columns=['relation'])
ff_data
```

```python
ff_data.to_csv('ff_data_2.txt', header=False, index=False)
```
